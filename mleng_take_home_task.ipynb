## 1. Baseline RAG — Implementation

Pipeline: chunk posting text → embed locally with `sentence-transformers` (no API key required) → index in two interchangeable vector stores (**FAISS** exact search and **turbovec** 4-bit quantized search) → retrieve top-k postings for a question → generate a grounded answer with Claude (`claude-haiku-4-5`) → return the answer with citations (job ID, title, company, location, excerpt).

If no Anthropic credentials are available (or a call fails), answer generation falls back to a deterministic, extractive mock answer built directly from the retrieved evidence, so the whole pipeline runs end-to-end without an API key.

````python
import anthropic

from jobmarket.config import (
    ANTHROPIC_MODEL,
    CHUNK_OVERLAP_WORDS,
    CHUNK_SIZE_WORDS,
    EMBEDDING_MODEL_NAME,
    TOP_K_CHUNKS,
    TOP_K_POSTINGS,
)

# Resolves ANTHROPIC_API_KEY if present; harmless if absent —
# absence/failure is handled at call time in generate_answer() below.
anthropic_client = anthropic.Anthropic()
````

### Chunk-size analysis

Before picking a chunk size, sanity-check it against the actual data rather than picking a round number blind.

**Why chunking is needed at all:** `description` (+ `skills_desc`) length varies hugely across postings — the distribution below shows a median around ~490 words, so most postings would be truncated (or leave most of their content unembedded) if each posting were indexed as a single vector.

**Why 180 words / 40-word overlap specifically:** the embedding model (`all-MiniLM-L6-v2`) has a hard **256-token** max sequence length — text beyond that is truncated silently by the tokenizer, not an error. 180 words is meant to sit comfortably under that limit, leaving headroom for the `Job title: ... / Company: ... / Location: ...` prefix that `build_corpus` prepends to every chunk before embedding (see `embedding_text` in [chunking.py](jobmarket/rag/chunking.py)). The 40-word overlap (~22% of the chunk) is a conventional fraction used to preserve context across chunk boundaries without materially inflating the number of chunks (and therefore index size / embedding cost).

````python
import numpy as np
from transformers import AutoTokenizer

from jobmarket.rag.chunking import split_into_chunks

MAX_SEQ_TOKENS = 256  # all-MiniLM-L6-v2's max_seq_length

# --- word-length distribution: does chunking make sense for this corpus at all? ---
body_text = (
    postings["description"].fillna("") + " " + postings["skills_desc"].fillna("")
).str.strip()
word_counts = body_text.str.split().apply(len)

print("description(+skills_desc) length in words:")
print(word_counts.describe(percentiles=[.25, .5, .75, .9, .95, .99]).round(1))
print(f"\nPostings longer than a single {CHUNK_SIZE_WORDS}-word chunk: "
      f"{(word_counts > CHUNK_SIZE_WORDS).mean():.1%}")

# --- token-budget check: word count is a proxy for token count, not the real constraint ---
tokenizer = AutoTokenizer.from_pretrained(EMBEDDING_MODEL_NAME)

sample_chunks = [
    chunk
    for text in body_text.sample(200, random_state=42)
    for chunk in split_into_chunks(text, CHUNK_SIZE_WORDS, CHUNK_OVERLAP_WORDS)
]
chunk_token_counts = np.array([len(tokenizer(c)["input_ids"]) for c in sample_chunks])

# embedding_text also prepends "Job title: ...\nCompany: ...\nLocation: ...\n\n" before the chunk text
prefix_tokens = len(tokenizer("Job title: X\nCompany: Y\nLocation: Z\n\n")["input_ids"])

print(f"\nTokenizer: {EMBEDDING_MODEL_NAME}, model max_seq_length={MAX_SEQ_TOKENS}")
print(f"Chunk token counts (n={len(chunk_token_counts)} chunks from {CHUNK_SIZE_WORDS}-word splitting): "
      f"mean={chunk_token_counts.mean():.0f}, p90={np.percentile(chunk_token_counts, 90):.0f}, "
      f"max={chunk_token_counts.max()}")
print(f"Chunks exceeding {MAX_SEQ_TOKENS} tokens (silently truncated by the model): "
      f"{(chunk_token_counts > MAX_SEQ_TOKENS).mean():.1%}")
print(f"Citation-prefix overhead added on top in embedding_text: ~{prefix_tokens} tokens")
````

**Measured on this dataset:** median `description`+`skills_desc` length is ~489 words (mean 532, up to 2,486) and 85% of postings exceed a single 180-word chunk, confirming multi-chunk splitting is necessary for most postings. Tokenizing 180-word chunks with the actual `all-MiniLM-L6-v2` tokenizer, however, shows word count is an imperfect proxy for token count: chunks average ~221 tokens (not 180), and roughly **1 in 5 chunks (19%) exceeds the 256-token limit** and has its tail silently truncated by the model — before even adding the ~12-token citation prefix.

This is accepted as a minor trade-off rather than fixed by shrinking the chunk size further: the 40-word overlap means a truncated tail is largely re-covered as the *start* of the next chunk, so no content is permanently lost from retrieval — only, occasionally, scored from a slightly shorter effective window. `CHUNK_SIZE_WORDS` / `CHUNK_OVERLAP_WORDS` stay at 180/40.

### Chunking

`description` (and `skills_desc`, where present) is unstructured free text of varying length. Each posting's text is split into overlapping ~180-word chunks (40-word overlap, rationale and token-budget analysis above) so long postings aren't truncated and retrieval can point at the relevant part of a posting rather than the whole document. Postings with no usable text are skipped. Each chunk keeps a back-reference to its parent posting (`job_id`, `title`, `company`, `location`) for citations.

````python
from jobmarket.rag.chunking import build_corpus, split_into_chunks

corpus = build_corpus(postings, chunk_size=CHUNK_SIZE_WORDS, overlap=CHUNK_OVERLAP_WORDS)
print(f"Built {len(corpus):,} chunks from {corpus['job_id'].nunique():,} postings "
      f"(out of {len(postings):,} total postings)")
````

A framework such as LangChain, provudes implementation for several chunking strategies. See: https://docs.langchain.com/oss/python/integrations/splitters 

### Embeddings

Chunks are embedded locally with `all-MiniLM-L6-v2` (384-dim, `sentence-transformers`) — no external API key required, so index-building always works offline. Embeddings are L2-normalized so inner-product search is equivalent to cosine similarity.

````python
from sentence_transformers import SentenceTransformer

embedder = SentenceTransformer(EMBEDDING_MODEL_NAME)

embeddings = embedder.encode(
    corpus["embedding_text"].tolist(),
    batch_size=64,
    show_progress_bar=True,
    normalize_embeddings=True,
    convert_to_numpy=True,
).astype("float32")

EMBEDDING_DIM = embeddings.shape[1]
print(f"Embedded {embeddings.shape[0]:,} chunks into {EMBEDDING_DIM}-dim vectors")
````

### Vector store: FAISS and turbovec

Two interchangeable local vector-search backends sit behind a common `VectorStore.search()` interface:

- **FAISS** (`IndexFlatIP`) — exact brute-force cosine search via normalized inner product. Simple and exact, the natural baseline at this corpus size (a few thousand chunks).
- **turbovec** (`IdMapIndex`, 4-bit quantized) — a SIMD, quantized index built for corpora far larger than this one (millions of vectors); included as a second approach, to compare retrieval quality/latency against the exact FAISS index at small scale, where the quantization trade-off doesn't pay for itself yet.

Note:
A framework like LangChain provides several vector store integration options.
URL: https://docs.langchain.com/oss/python/integrations/vectorstores

````python
from jobmarket.rag.vector_store import build_vector_stores

vector_stores = build_vector_stores(embeddings)
print("Vector stores ready:", list(vector_stores.keys()))
````

### Retrieval

A query is embedded with the same model, searched against the chosen backend, then chunks are mapped back to their source posting and de-duplicated by `job_id` (keeping each posting's best-scoring chunk) so citations point at distinct postings rather than repeating one posting multiple times.

````python
from jobmarket.rag.retrieval import retrieve as _retrieve


def retrieve(
    query: str,
    k_chunks: int = TOP_K_CHUNKS,
    k_postings: int = TOP_K_POSTINGS,
    backend: str = "faiss",
) -> list[dict]:
    return _retrieve(
        query,
        corpus=corpus,
        vector_stores=vector_stores,
        embedder=embedder,
        k_chunks=k_chunks,
        k_postings=k_postings,
        backend=backend,
    )
````

### Answer generation

Retrieved postings are formatted as a numbered evidence list and passed to Claude (`claude-haiku-4-5`, chosen for speed/cost during iterative development — see design note) with instructions to answer using only that evidence and to cite postings by their bracket number. If no Anthropic credentials are available, or the call fails for any reason (network, auth, rate limit), `generate_answer` falls back to a deterministic, extractive mock answer built directly from the retrieved excerpts — the "mock mode" called for by the assignment, so the pipeline always runs end-to-end without an API key.

````python
from jobmarket.rag.generation import SYSTEM_PROMPT, format_evidence, mock_answer
from jobmarket.rag.generation import generate_answer as _generate_answer


def generate_answer(query: str, evidence: list[dict], model: str = ANTHROPIC_MODEL) -> str:
    return _generate_answer(query, evidence, anthropic_client=anthropic_client, model=model)
````

````python
def answer_question(
    query: str,
    backend: str = "faiss",
    k_chunks: int = TOP_K_CHUNKS,
    k_postings: int = TOP_K_POSTINGS,
    model: str = ANTHROPIC_MODEL,
    verbose: bool = True,
) -> dict:
    evidence = retrieve(query, k_chunks=k_chunks, k_postings=k_postings, backend=backend)
    answer = generate_answer(query, evidence, model=model)

    if verbose:
        print(f"Q: {query}\n(backend={backend})\n")
        print(answer)
        print("\nCitations:")
        for i, e in enumerate(evidence, start=1):
            print(f"  [{i}] job_id={e['job_id']} | {e['title']} | {e['company']} | {e['location']} | score={e['score']:.3f}")
            print(f"      \"{e['excerpt'][:160].strip()}...\"")

    return {"query": query, "answer": answer, "citations": evidence, "backend": backend}
````

## 2. Knowledge Graph — Implementation

Built with **NetworkX** (`DiGraph`) — no server, easy to inspect in the notebook, and more than sufficient for a few thousand postings.

### Schema

| Node type | Key | Source |
|---|---|---|
| `JobPosting` | `job_id` | `sample_postings.csv` |
| `Company` | normalized `company_name` | `sample_postings.csv`, enriched from `sample_companies.csv` when `company_id` joins |
| `Role` | canonical role name | derived from `title` via a small keyword taxonomy |
| `Skill` | canonical skill name | extracted from `title` + `description` + `skills_desc` via keyword matching |
| `Location` | raw `location` string | `sample_postings.csv` |

Edges: `(JobPosting)-POSTED_BY->(Company)`, `(JobPosting)-HAS_ROLE->(Role)`, `(JobPosting)-MENTIONS_SKILL->(Skill)`, `(JobPosting)-LOCATED_IN->(Location)`.

**Deliberately not implemented:**
- `Industry` — neither `sample_postings.csv` nor `sample_companies.csv` carries an `industry_id` to join against `sample_industries.csv` in this sample, so there's no reliable edge to draw.
- `Occupation` (ESCO/O*NET) — optional enrichment, out of scope.
- `Skill-RELATED_TO-Skill` — not needed by the target questions; the required demo question is answered directly via posting-level traversal and set aggregation, no derived skill-similarity edges required.

**Key design trade-offs:**
- **Skills are *not* `sample_skills.csv`'s categories** (`IT`, `ENG`, `ANLS`, …) — those are too coarse to answer skill-level questions (noted during initial data exploration above). Instead, `Skill` nodes come from matching a curated vocabulary of ~45 concrete tech/data skills against posting text via case-insensitive keyword matching with symbol-aware boundaries (handles `C++`, `CI/CD`, `Node.js`, etc.). This is plain regex, not NER — fast and transparent, with an accepted false-negative/positive rate on ambiguous short tokens (documented at the vocabulary definition below).
- **`Role` is derived from `title`** using an ordered keyword taxonomy (most-specific pattern wins), covering the ~10 canonical roles that dominate this corpus (Data Engineer, Machine Learning Engineer, Data Scientist, DevOps Engineer, Software Engineer, …) — this directly supports the ML-Engineer-vs-Data-Scientist and data-engineering/ML-engineering-overlap questions.
- **`Company` is keyed by normalized name, not `company_id`**, since `company_id` is frequently `NaN` in the sample (per the earlier exploration cells) — this trades perfect de-duplication (name variants could fragment one real company into multiple nodes) for actually connecting most postings to a company node.
- **`Location` is the raw string, unnormalized** (e.g. "Boston, MA" and a hypothetical "Boston, Massachusetts" would be distinct nodes) — a scoping decision, not implemented here.

### Graph queries

Small traversal helpers over `kg`, composed to answer the required knowledge-graph question by **traversal** (posting → company, posting → skill) plus **aggregation** (set intersection across two role-filtered subgraphs) — no query language needed at this scale.

````python
from jobmarket.kg.queries import company_for_posting as _company_for_posting
from jobmarket.kg.queries import postings_by_company_for_role as _postings_by_company_for_role
from jobmarket.kg.queries import postings_for_role as _postings_for_role
from jobmarket.kg.queries import skills_for_posting as _skills_for_posting
from jobmarket.kg.queries import skills_for_role as _skills_for_role


def postings_for_role(role: str) -> list[str]:
    return _postings_for_role(kg, role)


def company_for_posting(posting_node: str) -> str | None:
    return _company_for_posting(kg, posting_node)


def skills_for_posting(posting_node: str) -> set[str]:
    return _skills_for_posting(kg, posting_node)


def postings_by_company_for_role(role: str) -> dict[str, list[str]]:
    return _postings_by_company_for_role(kg, role)


def skills_for_role(role: str) -> Counter:
    return _skills_for_role(kg, role)
````

## 3. Graph-RAG Design

*Architectural description only — not implemented. The knowledge graph built in Part 2 already provides everything a real implementation would traverse; this section describes how it would plug into the Part 1 RAG pipeline.*

### Where baseline RAG alone falls short

Evidence from the demos already run above:

- **"Given Python, SQL, Docker, and NLP experience, what adjacent roles are relevant?"** — the RAG-only answer surfaced a *"Senior Software Engineer-(Mainframe/SQL/ETL)"* posting alongside genuinely adjacent roles. Vector similarity matched on shared keywords (SQL, generic tech-stack overlap) with no notion of "how close is this role, structurally, to the input skill profile" — there's no set-based or frequency-based reasoning in embedding space.
- **"Which skills distinguish ML Engineer from Data Scientist roles?"** — the RAG-only answer returned five individually-relevant chunks, but a genuine *difference* between two role categories is a comparison of two full frequency distributions, not something a handful of retrieved excerpts can establish reliably.
- More generally: vector top-k is a **sample** (k≈5–8 chunks out of 16,378 in this corpus). Any question requiring exhaustive coverage — "which companies do X **and** Y," "what fraction of postings mention Z" — cannot be answered correctly by sampling, no matter how good the embedding model is.

### Which question types benefit from graph-aware retrieval, and why

| Question type | Example (from this assignment) | Best mechanism | Why |
|---|---|---|---|
| Multi-hop entity intersection | *"Which companies hire for both data engineering and ML engineering roles?"* | Graph traversal + set intersection (**implemented in Part 2**) | Needs two role-filtered subsets intersected by `Company` — vector similarity has no native intersection operator |
| Exhaustive aggregation / co-occurrence | *"What skills most often co-occur with LLM, generative AI, or RAG requirements?"* | Graph: shared-neighbor counting over `MENTIONS_SKILL` edges | Needs corpus-wide counts; `Skill` nodes already exist for `Large Language Models`, `Generative AI`, `Retrieval-Augmented Generation` — co-occurring skills are just the other `Skill` neighbors of the same `JobPosting` nodes |
| Structured skill-profile comparison | *"Which skills distinguish ML Engineer from Data Scientist roles?"* | Graph: per-`Role` skill-frequency profile, then relative-frequency diff | "Distinguishing" is a comparison of two aggregated distributions, not a similarity search |
| Nearest-role / skill-gap reasoning | *"Given Python, SQL, Docker, NLP, what adjacent roles are relevant and what's missing?"* | Graph: rank `Role`s by skill-profile overlap with the input set, diff the top role's profile against the input | A role's skill profile is a structured aggregate the graph already computes (`skills_for_role`); posting-level text search can't isolate that role-level signal from noise |
| Open-ended semantic / narrative discovery | *"Which postings mention retrieval-augmented generation, vector databases, or LLM application development?"* | Vector RAG (**as implemented in Part 1**) | Free phrasing, no fixed taxonomy required — exactly what embeddings are for |
| Grounded quoting for citations | any answer needing a supporting excerpt | Vector RAG / posting text (**as implemented in Part 1**) | The graph tells you *which* postings qualify; only the source text supplies a quotable excerpt |

Most real questions are a mix of the top and bottom halves of this table — which is why the proposed design **combines** graph traversal and vector retrieval rather than replacing one with the other.

### Proposed architecture

```text
                       ┌───────────────────────┐
  natural-language ───▶│ 1. Query parsing       │
  question             │   role/skill/company/  │
                       │   location + intent op │
                       └───────────┬────────────┘
                                   │ structured intent
                                   ▼
                       ┌───────────────────────┐
                       │ 2. Graph traversal     │
                       │   (kg: exact,          │
                       │    exhaustive)         │
                       └───────────┬────────────┘
                candidate job_ids + aggregate stats
                       + traversal provenance
                                   │
                                   ▼
                       ┌───────────────────────┐
                       │ 3. Vector retrieval    │
                       │   restricted/boosted   │
                       │   to the candidate set │
                       └───────────┬────────────┘
                        quotable excerpts, scored
                                   │
                                   ▼
                       ┌───────────────────────┐
                       │ 4. Answer generation   │
                       │   (graph facts +       │
                       │    vector excerpts)    │
                       └───────────────────────┘
```

**Stage 1 — Query parsing.** Reuse `classify_role` / `extract_skills` (already built in Part 2) plus lightweight entity matching for company/location names to turn the question into a structured intent: `{roles: [...], skills: [...], companies: [...], locations: [...], op: "intersect" | "co_occur" | "diff" | "rank" | "open"}`. An LLM call (same model as answer generation) is the pragmatic way to classify `op` and resolve entities that don't match the fixed vocabularies exactly (e.g. "ML engineering roles" → `Role: Machine Learning Engineer`).

**Stage 2 — Graph traversal.** Run the structured intent against `kg` using the same traversal primitives from Part 2 (`postings_for_role`, `postings_by_company_for_role`, `skills_for_role`, `companies_bridging_roles`) or natural extensions of them (e.g. a `skill_cooccurrence(skill)` that walks `MENTIONS_SKILL` backward to `JobPosting` then forward to other `Skill`s). Output:
- a **candidate `job_id` set** (exact, not sampled) to hand to vector search;
- **aggregate statistics** computed over the whole graph (counts, intersections, differences) — not approximated from k retrieved chunks;
- **traversal provenance** — which edges justified each result (e.g. "Capital One qualifies via 22 `Data Engineer` postings and 5 `Machine Learning Engineer` postings") — citation-quality explainability a similarity score alone can't provide.

**Stage 3 — Graph-guided vector retrieval.** Instead of searching the full ~16K-chunk index, restrict `retrieve()` to the graph-derived candidate `job_id`s — an *allowlist* filter, the same hybrid pattern `turbovec`'s `IdMapIndex.search()` already supports natively (candidates narrowed by an external system, then ranked by embedding similarity within that set). This gets the best of both: exact structural correctness from the graph, plus a natural-language excerpt worth quoting from the vector index. Where the graph traversal alone already answers the question numerically (counts, company lists), this stage's only job is finding *one good supporting quote per fact*, not re-discovering which postings are relevant.

**Stage 4 — Answer generation.** Same `generate_answer()` from Part 1, but the evidence block passed to the LLM now has two tiers: **graph facts** (exact counts, set memberships, comparisons — computed, not estimated) and **vector excerpts** (quotable text, tied to specific `job_id`s for citation). This lets the model make statistically-grounded claims ("14 companies post both role types; Python and SQL are the two most common shared skills, appearing in 9 of them") backed by a real excerpt, instead of inferring a pattern from 5 retrieved snippets.

### What the graph step contributes

**Input:** the structured intent from Stage 1 (roles/skills/companies/locations + operation type).

**Output:**
1. A candidate `job_id` set — exact and exhaustive, not a top-k sample — that narrows and improves the precision of the vector-search stage.
2. Aggregate statistics (frequency counts, set intersections/differences, co-occurrence) that vector similarity cannot compute at all.
3. Traversal provenance for each result, giving the answer generator (and the eventual reader) an auditable "why" beyond a similarity score.

**Signals that improve answer quality specifically:**
- **Exhaustiveness** — corpus-wide counts instead of a k-sized sample, directly fixing the "distinguishing skills" and "co-occurrence" question types.
- **Structural operations** (intersection, difference, ranking) that have no equivalent in embedding space.
- **Canonicalization** — `Role` and `Skill` nodes collapse title/skill variants (e.g. "ML Engineer", "Machine Learning Engineer", "Senior Machine Learning Engineer" → one `Role` node) that vector search would otherwise treat as separate, loosely-related points in embedding space.
- **Provenance** — a traversal path is a stronger citation than a cosine-similarity score.

### Concrete walkthrough

**"Which companies hire for both data engineering and ML engineering roles, and what skills connect those roles?"** — the required demo question, already answered purely by graph traversal in Part 2. A full Graph-RAG pipeline would improve on that pure-graph answer in one specific way: **Stage 3** would attach a real supporting quote to each bridging company — e.g. for Capital One, retrieve the single best-matching excerpt from its Data Engineer and ML Engineer postings that best evidences "Python" or "Spark" — turning the current company/skill list into a fully cited answer, the same citation quality as the Part 1 RAG demos.

**"Given Python, SQL, Docker, and NLP experience, what adjacent roles are relevant, and which skills are most commonly missing?"** shows the gap more starkly. The Part 1 RAG-only demo surfaced a *Mainframe/SQL/ETL* posting as a top match — a false positive driven by superficial keyword overlap. A graph-aware version would instead: (1) build the input skill set `{Python, SQL, Docker, NLP}`; (2) for every `Role`, compute `skills_for_role(role)` and rank roles by overlap (e.g. Jaccard similarity) with the input set — `Machine Learning Engineer` and `Data Engineer` would score highest, generic `Software Engineer` moderately, and the Mainframe posting's actual role would rank far lower once measured against its *full* skill profile rather than one chunk's text; (3) diff the top-ranked role's aggregate skill profile against the input set to report the *missing* skills (e.g. "Machine Learning Engineer postings in this corpus also commonly require: AWS, Kubernetes, Spark — Docker is already covered"); (4) hand the top 1–2 qualifying postings per recommended role to vector retrieval for a quotable excerpt. This directly fixes the specific failure mode observed in the Part 1 demo.
